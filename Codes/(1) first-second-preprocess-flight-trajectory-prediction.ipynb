```python
import requests
from bs4 import BeautifulSoup

def is_helicopter(callsign):
    url = f"https://www.flightaware.com/resources/registration/{callsign}"

    # Create a session to persist certain parameters across requests
    session = requests.Session()

    # Set the headers to mimic a browser request
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
        'Accept-Language': 'en-US,en;q=0.5',
        'Referer': 'https://www.google.com',
        'Connection': 'keep-alive',
        'Upgrade-Insecure-Requests': '1'
    }

    # Make the HTTP request with headers
    response = session.get(url, headers=headers)

    # Check the HTTP response status
    if response.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find the specific section containing "Rotorcraft"
        aircraft_summary = soup.find("div", {"class": "medium-3 columns"})

        # Check if "Rotorcraft" is in the content of the specific section
        if aircraft_summary and "rotorcraft" in aircraft_summary.get_text().lower():
            return True
    return False

# Example usage
callsign = "N464GR"
print(is_helicopter(callsign))  # Prints True if it's a helicopter, otherwise False
```

```python
# Check each callsign and determine if it represents a helicopter
helicopter_callsigns = []
loop_count = 0

for callsign in n_callsigns:
    loop_count += 1
    if is_helicopter(callsign):
        helicopter_callsigns.append(callsign)

    print(f"Loop {loop_count}: Checked callsign {callsign}")

print(f"Number of helicopters: {len(helicopter_callsigns)}")
print("Helicopter callsigns:", helicopter_callsigns)
```

```python
heli = helicopter_callsigns
print(heli)
heli_df = pd.DataFrame(heli, columns=['Helicopter_Callsigns'])
heli_df.to_csv('helicopter_callsigns.csv', index=False)
```
