```python
import pandas as pd
import matplotlib.pyplot as plt
import re

df = pd.read_csv("/kaggle/input/flight-trajectory-dataset/9columns_and_nullcallsign_dropped_flights_data.csv")
df
```

```python
heli_df = pd.read_csv('/kaggle/input/flight-trajectory-dataset/helicopter_callsigns.csv')
heli_callsigns = heli_df['Helicopter_Callsigns'].values
heli_callsigns
```

```python
# Remove rows where the 'callsign' column contains any value in the 'heli' list
df_noheli = df[~df['callsign'].isin(heli_callsigns)]
df_noheli = df_noheli[df_noheli['time'] % 2 == 0]
df_noheli = df_noheli.drop(columns=['heading', 'baroaltitude'])

print("\nNon-helicopter DataFrame:")
print(df_noheli)
```

```python
df_noheli.to_csv('/kaggle/working/non_helicopter_7columns_2s_flights_data.csv', index=False)
```

```python
df_1 = pd.read_csv("/kaggle/working/non_helicopter_7columns_2s_flights_data.csv")
df_1
```

```python
df_2 = df_1.copy()
# Convert the 'time' column from timestamp to human-readable date
df_2['human_date'] = pd.to_datetime(df_2['time'], unit='s')

# Reorder columns to place 'human_date' right next to 'time'
columns = df_2.columns.tolist()
time_index = columns.index('time')
columns.insert(time_index + 1, columns.pop(columns.index('human_date')))
df_2 = df_2[columns]
df_2
```

```python
df_3 = df_2.copy()
# Calculate the time difference between consecutive rows
df_3['time_diff'] = df_3['time'].diff()

# Set a threshold for the maximum duration between two consecutive data points within the same flight segment
threshold = 30  # 30 secs

# Identify points where the time difference exceeds the threshold
df_3['new_segment'] = (df_3['time_diff'] > threshold).cumsum()

# Split the dataframe into separate flight segments
flight_segments_time_based = {name: group for name, group in df_3.groupby('new_segment')}

# Display the information about the flight segments based on time threshold
flight_segment_time_info = pd.DataFrame({
    'Segment': list(flight_segments_time_based.keys()),
    'Start Time': [group['time'].min() for group in flight_segments_time_based.values()],
    'End Time': [group['time'].max() for group in flight_segments_time_based.values()],
    'Duration (mins)': [(group['time'].max() - group['time'].min()) / 60 for group in flight_segments_time_based.values()]
})

# Show the DataFrame with flight segment information
flight_segment_time_info
```

```python
flight_segment_time_info_sorted = flight_segment_time_info.sort_values(by='Duration (mins)', ascending=False)

# Display the sorted DataFrame
print(flight_segment_time_info_sorted)
```

```python
# Find the number of segments with a duration shorter than 1 minute or longer than 30 mintues
segments_1_30min = flight_segment_time_info[(flight_segment_time_info['Duration (mins)'] < 1) | (flight_segment_time_info['Duration (mins)'] > 30)]

# Number of such segments
num_segments_1_30min = len(segments_1_30min)

# Display the count
print("Number of segments with duration shorter than 1 minute or longer than 30 minutes:", num_segments_1_30min)
```

```python
# Filter the segments with 'Duration (mins)' smaller than 1 or greater than 30
filtered_segments = flight_segment_time_info_sorted[
    (flight_segment_time_info_sorted['Duration (mins)'] < 1) | 
    (flight_segment_time_info_sorted['Duration (mins)'] > 30)
]

filtered_segments
```

```python
# Initialize an empty list to collect the need-to-remove 'time' values
remove_times = []
loop_count = 0

# Iterate over each filtered segment and collect the 'time' values
for index, row in filtered_segments.iterrows():
    loop_count += 1
    start_time = row['Start Time']
    end_time = row['End Time']
    segment_times = df_1[(df_1['time'] >= start_time) & (df_1['time'] <= end_time)]['time'].values
    remove_times.extend(segment_times)
    if (loop_count) % 1000 == 0:
        print(f"Loop {loop_count}")
```

```python
import numpy as np

# Convert the list to a numpy array
remove_times_array = np.array(remove_times)

# Remove duplicates to get unique values
unique_remove_times_array = np.unique(remove_times_array)

# Print the array and its length
print(unique_remove_times_array)
print(len(unique_remove_times_array))
```

```python
remove_times_df = pd.DataFrame(unique_remove_times_array, columns=['Removed_times_1_30'])
remove_times_df.to_csv('/kaggle/working/removed_times_1_30.csv', index=False)
```

```python
# Filter the original DataFrame to exclude these times
df_1_30min_removed = df_1[~df_1['time'].isin(unique_remove_times_array)]

df_1_30min_removed
```

```python
df_1_30min_removed.to_csv('/kaggle/working/non_helicopter_7columns_2s_remove_1_30mins_flights_data.csv', index=False)
```
